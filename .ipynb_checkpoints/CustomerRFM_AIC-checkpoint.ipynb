```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_excel("Online_Retail.xlsx")
df.head()
import datetime as dt
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']).dt.date
df = df[pd.notnull(df['CustomerID'])]
df = df[(df['Quantity']>0)]
df['Sales'] = df['Quantity'] * df['UnitPrice']
cols_of_interest = ['CustomerID', 'InvoiceDate', 'Sales']
df = df[cols_of_interest]
print(df.head())
print(df.head())
print(df['CustomerID'].nunique())
```

```python
from lifetimes.plotting import *
from lifetimes.utils import *
from sklearn import preprocessing

data = summary_data_from_transaction_data(df, 'CustomerID', 'InvoiceDate', monetary_value_col='Sales', observation_period_end='2011-12-9')
data.head()
```

```python
data = data[['frequency', 'recency', 'monetary_value']]
data.head()
```

```python
X = preprocessing.normalize(data)
```

```python
X = pd.DataFrame(X)
X.columns = ['frequency', 'recency', 'monetary_value']
```

```python
from sklearn.cluster import KMeans 
from sklearn import metrics 
from scipy.spatial.distance import cdist 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
K = range(1,10) 
  
for k in K: 
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=k).fit(X) 
    kmeanModel.fit(X)     
      
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 
                      'euclidean'),axis=1)) / X.shape[0]) 
    inertias.append(kmeanModel.inertia_) 
  
    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_, 
                 'euclidean'),axis=1)) / X.shape[0] 
    mapping2[k] = kmeanModel.inertia_ 
```

```python
for key,val in mapping1.items(): 
    print(str(key)+' : '+str(val))
```

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show() 
```

```python
for key,val in mapping2.items(): 
    print(str(key)+' : '+str(val)) 
```

```python
plt.plot(K, inertias, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show() 
```

```python
knn = KMeans(n_clusters=3).fit(X) 
X_knn = knn.fit_predict(X)
```

```python
X['cluster'] = X_knn
```

```python
X.head()
```

```python
plt.scatter(X['frequency'], X['recency'], label=X['cluster'], c=X['cluster'])
plt.xlabel('Frequency')
plt.ylabel('Recency')
plt.title('Customer Pattern (Frequency vs Recency)')
plt.show()
```

```python
plt.scatter(X['monetary_value'], X['recency'], c=X['cluster'], alpha=0.5)
plt.xlabel('monetary_value')
plt.ylabel('Recency')
plt.title('Customer Pattern (Monetary Value vs Recency)')
```

```python
plt.scatter(X['monetary_value'], X['frequency'], c=X['cluster'], alpha=0.5)
plt.xlabel('monetary_value')
plt.ylabel('frequency')
plt.title('Customer Pattern (Monetary Value vs Frequency)')
```
